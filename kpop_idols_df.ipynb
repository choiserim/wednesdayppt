```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
```

```python
idols_df = pd.read_csv('/content/drive/MyDrive/data/kpop_idols.csv')
idols_df
```

```python
idols_df.info()
```

```python
idols_df.describe()
```

```python
girl_group = idols_df.query('Gender == "Girl"')
boy_group = idols_df.query('Gender == "Boy"')

y_mean = idols_df['Followers'].mean()
```

```python
fig, ax = plt.subplots()

ax.scatter(girl_group['Career'], girl_group['Followers'],  marker='*', alpha=0.5, color='coral', label='girl group')
ax.scatter(boy_group['Career'], boy_group['Followers'], marker='p', alpha=0.5, color='lightseagreen', label='boy group')

# ax.axhline(y=y_mean, color='gray', linestyle='--', linewidth=0.5, label='mean')
ax.axhline(idols_df['Followers'].mean(), color='grey', linestyle='--', linewidth=0.5, label='mean')

ax.set_title('Followers per Career Length')  # 제목 붙이기
ax.set_xlabel('Years of Career') # x축 라벨 붙이기
ax.set_ylabel('Instagram Followers (in Millions)')  # 단위에 대한 설명 추가
ax.set_xticks(np.arange(0, 18, 1))
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: int(x / 1e6)))  # y축 단위를 Million 단위로 맞추기

# ax.spines.top.set_visible(False)
# ax.spines.right.set_visible(False)
ax.spines[['right', 'top']].set_visible(False)

# 마지막으로 그래프에 범례(legend)를 붙여주세요. 평균선의 라벨까지 범례에 포함되도록 하려면 범례를 생성하는 코드를 ax.axhline() 보다 뒤에 넣어야 한다는 점을 유의해 주세요.
ax.legend()
plt.show()
```
